```python
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')
data['Weight'] = data['Weight'].str.replace('kg','').astype('float32')
```

```python
def add_company(inpt):
    if inpt == 'Samsung' or inpt == 'Razer' or inpt == 'Mediacom' or inpt == 'Microsoft'or inpt == 'Xiaomi'or inpt == 'Vero'or inpt == 'Chuwi'or inpt == 'Google'or inpt == 'Fujitsu'or inpt == 'LG'or inpt == 'Huawei':
        return 'Other'
    else:
        return inpt
```

```python
# modified ‘Company’ column with updated company names based on the function’s logic.
data['Company'] = data['Company'].apply(add_company)
```

```python
data['Company'].value_counts()
```

```python
len(data['Product'].value_counts())
```

```python
data['TypeName'].value_counts()
```

```python
data['ScreenResolution'].value_counts()
```

```python
# The lambda function checks if ‘IPS’ is present in the ‘ScreenResolution’.
# If yes, it assigns 1; otherwise, it assigns 0.
# This column indicates whether a laptop has an IPS (In-Plane Switching) display.
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head(10)
```

```python
data['Cpu'].value_counts()
```

```python
# The lambda function takes each value in the ‘Cpu’ column.
# It splits the value into words using spaces (x.split()).
# Then, it selects the first three words (e.g., brand, series, and model) using slicing (x.split()[0:3]).
# Finally, it joins these three words back together with spaces (" ".join(...)
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))

# DataFrame now has a ‘cpu_name’ column containing the concatenated brand, series, and model information from the original ‘Cpu’ column.
```

```python
data['cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'
```

```python
data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['cpu_name'].value_counts()
```

```python
data['Ram'].value_counts()
```

```python
data['Gpu'].value_counts()
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
data['gpu_name'].value_counts()
```

```python
data.shape
```

```python
data.head(5)
```

```python
data = data[data['gpu_name'] != 'ARM']
data.shape
```

```python
data.head(5)
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return inpt
    else:
        return 'Other'
    
data['OpSys'] = data['OpSys'].apply(set_os)
data['OpSys'].value_counts()
```

```python
data.head(5)
```

```python
data = data.drop(columns=['laptop_ID', 'Inches', 'Product', 'ScreenResolution', 'Cpu', 'Gpu'])
data
```

```python
data = pd.get_dummies(data)
```

```python
data.head(5)
```

```python
data.shape
```

Convert dataframe into dependent variable and independent variable

```python
x = data.drop('Price_euros',axis=1)
y = data['Price_euros']
```

# Model Selection

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
x_train.shape, x_test.shape
```

```python
# The function prints the model’s name (converted to a string) along with its accuracy score.

def model_acc(model):
    model.fit(x_train, y_train)
    acc = model.score(x_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

# Model Testing

 The **RandomForestRegressor** is a powerful machine learning algorithm. Let's explore its key features:

1. **What Is It?**:
   - The **RandomForestRegressor** is part of the scikit-learn library.
   - It's a **meta estimator** that combines multiple decision tree regressors.
   - Each tree is trained on a different subset of the dataset.
   - Averaging the predictions from these trees improves predictive accuracy and helps control overfitting¹.

2. **How It Works**:
   - Random forests create an ensemble of decision trees.
   - Each tree is trained on a random subset of features and data points.
   - The final prediction is an average (or weighted average) of individual tree predictions.

3. **Parameters**:
   - `n_estimators`: Number of trees in the forest.
   - `criterion`: Quality measure for splits (e.g., mean squared error, mean absolute error, etc.).
   - `max_depth`: Maximum depth of each tree.
   - `min_samples_split`: Minimum samples required to split an internal node.
   - `min_samples_leaf`: Minimum samples required at a leaf node.

4. **Use Cases**:
   - Regression tasks (predicting numerical values).
   - Robustness against overfitting.
   - Handling missing data and noisy features.





```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```
